```python
"""
4.2 画图让抽象问题形象化
面试题19：求二叉树的镜像
思路一：按层次遍历，每一层从右到左
思路二：递归遍历
"""
from collections import deque

class TreeNode(object):
    def __init__(self, x):
        self.val = x
        self.left = None
        self.right = None
        
class Tree(object):
    def __init__(self):
        self.root = None
        
    def construct_tree(self, values=None):
        if not values:
            return None
        self.root = TreeNode(values[0])
        queue = deque([self.root])
        leng = len(values)
        nums = 1
        while nums < leng:
            node = queue.popleft()
            if node:
                node.left = TreeNode(values[nums]) if values[nums] else None
                queue.append(node.left)
                if nums + 1 < leng:
                    node.right = TreeNode(values[nums + 1]) if values[nums + 1] else None
                    queue.append(node.right)
                    nums += 1
                nums += 1
                
    def bfs(self):
        ret = []
        queue = deque([self.root])
        while queue:
            node = queue.popleft()
            if node:
                ret.append(node.val)
                queue.append(node.left)
                queue.append(node.right)
        return ret
    
    def pre_traversal(self):
        ret = []
        def traversal(head):
            if not head:
                return
            ret.append(head.val)
            traversal(head.left)
            traversal(head.right)
        traversal(self.root)
        return ret
    
    def in_traversal(self):
        ret = []
        def traversal(head):
            if not head:
                return
            ret.append(head.left)
            traversal(head.val)
            traversal(head.right)
        traversal(self.root)
        return ret
    
    def post_traversal(self):
        ret = []
        def traversal(head):
            if not head:
                return
            ret.append(head.left)
            traversal(head.right)
            traversal(head.val)
        traversal(self.root)
        return ret
    
def mirror_bfs(root):
    ret = []
    queue = deque([root])
    while queue:
        node = queue.popleft()
        if node:
            ret.append(node.val)
            queue.append(node.right)
            queue.append(node.left)
    return ret

def mirror_pre(root):
    ret = []
    def traversal(head):
        if not head:
            return 
        ret.append(head.val)
        traversal(head.right)
        traversal(head.left)
    traversal(root)
    return ret


if __name__ == '__main__':
    t = Tree()
    t.construct_tree([1, 3, 2, 7, 6, 5, 4])
    print(t.bfs())
    print(mirror_bfs(t.root))
    print(mirror_pre(t.root))
```

```python
"""
4.2 画图让抽象问题形象化
面试题20：按从外到里的顺序顺时针打印矩阵
每一圈的开始位置总是坐上角元素[0, 0], [1, 1]...
"""
# 力扣方法：https://leetcode-cn.com/problems/shun-shi-zhen-da-yin-ju-zhen-lcof/solution/yi-xing-dai-ma-python3-yun-yong-ziphe-jin-xing-ju-/
def likou_func(matrix):
    """
    通过旋转矩阵的方式，每次将待遍历的数字放到最上一层列表里
    """
    res = []
    while matrix:
        res += matrix.pop(0)
        print(f"res===>{res}")
        matrix = list(zip(*matrix))[::-1]
        print(f"matrix===>{matrix}")
    return res

# 力扣精选方法-模拟法
# https://leetcode-cn.com/problems/shun-shi-zhen-da-yin-ju-zhen-lcof/solution/mian-shi-ti-29-shun-shi-zhen-da-yin-ju-zhen-she-di/
def print_matrix(matrix):
    if not matrix:
        return []
    l, r, t, b, res = 0, len(matrix[0]) - 1, 0, len(matrix) - 1, []
    while 1:
        # left -> right
        for i in range(l, r+1):  # 这个是 l ==> r+1
            res.append(matrix[t][i])
        t += 1
        if t > b: break  # t 是标记起点，大于 b，即大于 总行数，即到头了
            
        # top -> bottom
        for i in range(t, b+1):
            res.append(matrix[i][r])
        r -= 1
        if l > r: break  # l 是标记起点，大于 r，即大于 总列数，即到头了
            
        # right -> left
        for i in range(r, l - 1, -1):
            res.append(matrix[b][i])
        b -= 1
        if t > b: break
    
        # bottom -> top
        for i in range(b, t - 1, -1):
            res.append(matrix[i][l])
        l += 1
        if l > r: break
    return res
    

if __name__ == '__main__':
    """
    mat = [[1, 2, 3],
           [5, 6, 7],
           [9, 10, 11]]
    mat = [[]]
    mat = [[1]]
    mat = [[1, 2, 3, 4]]
    mat = [[1], [2], [3], [4]]
    """
#     mat = [[1, 2],
#            [5, 6]]
    mat = [[1, 2, 3],
           [5, 6, 7],
           [9, 10, 11]]
#     print_matrix(mat)
#     print(likou_func(mat))  # [1, 2, 3, 7, 11, 10, 9, 5, 6]
    print(print_matrix(mat))
```

```python
"""
4.3 举例让抽象问题具体化
面试题21：包含min函数的栈
栈的push，pop，min操作的时间复杂度都是O(1)
使用一个辅助栈保存最小值
https://leetcode-cn.com/problems/bao-han-minhan-shu-de-zhan-lcof/solution/mian-shi-ti-30-bao-han-minhan-shu-de-zhan-fu-zhu-z/

解析：
    将输入 item 压入栈，如果 min 为空，或者 min 的倒数第一个值 小于 item， 那么 min 添加 item
    
    抛出 stack 最后一个，如果抛出的这个，正好是 min 的倒数第一个，那么 min 也抛出最后一个
    
    min 始终存的是 stack 中的最小值，且最小值一直在 栈顶。
"""
class MyStack(object):
    
    def __init__(self):
        self.stack = []
        self.min = []
        
    def push(self, item):
        self.stack.append(item)
        if not self.min or item < self.min[-1]:
            self.min.append(item)
    
    def pop(self):
        if self.stack.pop() == self.min[-1]:
            self.min.pop()
    
    def min(self):
        return self.min[-1] if self.min else None
    
if __name__ == '__main__':
    s = MyStack()
    s.push(2)
    s.push(1)
    s.push(3)
    s.pop()
    s.push(2)
    print(s.stack)
    print(s.min)
```

```python
"""
4.3 举例让抽象问题具体化
面试题22：栈的压入弹出序列，判断给定的两个序列中，后者是不是前者的弹出序列，序列中不存在相同值
使用一个辅助栈, 如果辅助栈栈顶元素不等于出栈元素，则从入栈中找改值，直到入栈为空
如果最后出栈序列为空，则是入栈的弹出序列

https://leetcode-cn.com/problems/zhan-de-ya-ru-dan-chu-xu-lie-lcof/solution/mian-shi-ti-31-zhan-de-ya-ru-dan-chu-xu-lie-mo-n-2/
"""
class MyStack(object):
    def validateStackSequences(self, pushed, popped):
        stack, i = [], 0   # 生成一个模拟栈，i 是指向 抛出序列的索引
        for num in pushed:
            stack.append(num)   # 入栈
            while stack and stack[-1] == popped[i]:  # 循环判断与出栈
                print(num, "--->", stack, "---->", i, "---->", stack[-1], "--->", popped[i])
                stack.pop()
                i += 1
        return not stack
    
if __name__ == '__main__':
    mystack = MyStack()
    print(mystack.validateStackSequences([1, 2, 3, 4, 5], [4, 5, 3, 2, 1]))
            
```

```python
"""
4.3 举例让抽象问题具体化
面试题24：判断给定的整数数组是不是二叉搜索树的后序遍历序列
题目地址：https://leetcode-cn.com/problems/er-cha-sou-suo-shu-de-hou-xu-bian-li-xu-lie-lcof/

后序遍历：左子树，右子树 根节点 即遍历顺序为 左、右、根
二叉搜索树：左子树中所有节点的值 < 根节点的值，右子树中所有节点的值 > 根节点的值；其左、右子树也分别为二叉搜索树
"""

class Soulution:
    def verifyPostorder(self, postorder):
        def recur(i, j):       # 说明节点数量 <= 1
            if i >= j: 
                return True
            p = i
            while postorder[p] < postorder[j]:
                p += 1
            m = p
            while postorder[p] > postorder[j]:
                p += 1
            return p == j and recur(i, m - 1) and recur(m, j - 1)
        return recur(0, len(postorder) - 1)
    
    
if __name__ == '__main__':
    postorder = [1, 3, 6, 2, 5]
    res = Soulution().verifyPostorder(postorder)
    print(res)
```

```python
"""
4.3 举例让抽象问题具体化
面试题25：二叉树中和为某一值的路径
题目地址：https://leetcode-cn.com/problems/er-cha-shu-zhong-he-wei-mou-yi-zhi-de-lu-jing-lcof/

思路：
1. 创建两个列表，一个存所有路径，一个存当前符合的路径
2. 如果没有根节点，返回
3. 添加当前根节点的值，目标值减去 当前根节点的值
4. 如果目标值以为0，
"""

class Solution:
    def pathSum(self, root, sum):
        res, path = [], []
        def recur(root, tar):
            if not root: 
                return
            path.append(root.val)  # 添加 path 中记录的数值
            tar -= root.val
            if tar == 0 and not root.left and not root.right:
                res.append(list(path))
            recur(root.left, tar)
            recur(root.right, tar)
            path.pop()   # 每次遍历保证 path 中为空

        recur(root, sum)
        return res
```

```python
"""
4.4 分解让复杂问题简单化
面试题27 二叉搜索树与双向链表
题目描述：输入一棵二叉搜索树，将该二叉搜索树转换成一个排序的循环双向链表。要求不能创建任何新的节点，只能调整树中节点指针的指向。
题目链接：https://leetcode-cn.com/problems/er-cha-sou-suo-shu-yu-shuang-xiang-lian-biao-lcof/
"""
class Solution:
    def treeToDoublyList(self, root: 'Node') -> 'Node':
        def in_traversal(cur):
            if not cur:
                return
            in_traversal(cur.left)  # 递归左子树
            if self.pre:
                self.pre.right, cur.left = cur, self.pre
            else:
                self.head = cur
            self.pre = cur
            in_traversal(cur.right)
        
        if not root:
            return
        self.pre = None
        in_traversal(root)
        self.head.left, self.pre.right = self.pre, self.head
```
