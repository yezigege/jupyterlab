```python
"""
面试题15 求链表中倒数第k个结点
思路: 使用快慢指针，快的先走k-1步，需要考虑控链表以及k为0
"""
# 方法一：遍历链表，存为列表，再行查找
def fun01(link, k):
    if not link or k<=0:  # 链表为空或倒数小于0
        return None
    
    stack = []
    cur = link
    while cur:            # 遍历链表，存放为栈
        stack.append(cur.val)
        cur = cur.next
        
    if k > len(stack):    # 所求大于链表长度
        return None
    
    return stack[-k]
    

# 方法二：快慢指针 两个指针，均指链头，A先k-1，B先不动。同时出发，A到尾部，B即结果
def lask_kth(link, k):
    if not link or k <= 0:
        return None
    
    move = link
    while move and k-1 >= 0:
        move = move.next
        k -= 1
    
    while move:
        move = move.next
        link = link.next
        
    if k == 0:
        return link.val
    return None

class ListNode(object):
    
    def __init__(self, val):
        self.val = val
        self.next = None
        
class Nodes(object):
    
    def __init__(self, values=None):
        self.nodes = self._set_link(values) if values else None
    
    def _set_link(self, values):
        head = ListNode(values[0])
        move = head
        try:
            for i in range(1, len(values)):
                tmp = ListNode(values[i])
                move.next = tmp
                move = move.next
        except Exception as e:
            print(e)
        return head.next
    
    def get_link(self):
        return self.nodes

if __name__ == '__main__':
    nodes = Nodes([1, 2, 3, 4, 5, 6, 7, 8, 9, 13])
    link = nodes.get_link()
    print(lask_kth(link, 4))
```

```python
"""
面试题16：反转链表
思路：需要考虑空链表，只有一个结点的链表, 参考leetcode上题解：https://leetcode-cn.com/problems/reverse-linked-list/solution/tu-jie-liu-cheng-python3die-dai-xiang-jie-by-han-h/
"""
def reverse_link(head):
    if not head or not head.next:
        return head
    cur, pre = head, None
    while cur:
        pre, pre.next, cur = cur, pre, cur.next  # 同时转换标签
    return pre

class LinkNode(object):
    
    def __init__(self, x):
        self.val = x
        self.next = None
        
class Nodes(object):
    
    def __init__(self, values=None):
        self.nodes = self._set_link(values) if values else None
        
    def get_link(self):
        return self.nodes
    
    def _set_link(self, values):
        head = LinkNode(0)
        move = head
        try:
            for val in values:
                temp = LinkNode(val)
                move.next = temp
                move = move.next
        except Exception as e:
            print(e)
        return head.next
    
    @staticmethod
    def print_link(link=None):
        count = 1
        while link:
            if count == 1:
                print(link.val)
            elif count % 5 == 0:
                print('--->', link.val)
            else:
                print('->', link.val)
            count += 1
            link = link.next
        print()
        
    def print_self(self):
        Nodes.print_link(self.nodes)
        
        
if __name__ == '__main__':
    nodes = Nodes([1, 2, 3, 4, 5, 6, 7, 8, 9, 13])
    link = nodes.get_link()
    nodes.print_self()
    head = reverse_link(link)  # 返回反转链表
    nodes.nodes = head
    print("反转后链表")
    Nodes.print_link(head)     # 1： 打印反转链表
    nodes.print_self()         # 2：同 1
```

```python
"""
面试题17：合并两个排序的链表
思路：使用递归
"""
def merge_link(head1, head2):
    if not head1:
        return head2
    if not head2:
        return head1
    if head1.val < head2.val:
        ret = head1
        ret.next = merge_link(head1.next, head2)
    else:
        ret = head2
        ret.next = merge_link(head1, head2.next)
    return ret

class ListNode(object):
    
    def __init__(self, x):
        self.val = x
        self.next = None
        
class Nodes(object):
    
    def __init__(self, values):
        self.nodes = self._set_link(values) if values else None
        
    def _set_link(self, values):
        head = LinkNode(None)
        move = head
        try:
            for val in values:
                temp = ListNode(val)
                move.next = temp
                move = move.next
        except Exception as e:
            print(e)
            
        return head.next

    def get_link(self):
        return self.nodes
    
    def print_self(self):
        Nodes.print_link(self.nodes)
        
    @staticmethod
    def print_link(link=None):
        if not link:
            return None
        while link:
            print(f'==>{link.val}')
            link = link.next
        print()
    

if __name__ == '__main__':
    nodes = Nodes([2, 4, 6, 8 ,10])
    h1 = nodes.get_link()
    h2 = Nodes([1, 3, 5, 7, 9]).get_link()
    ret = merge_link(h1, h2)
    Nodes.print_link(ret)
```

```python
"""
面试题18：判断一棵二叉树是不是另一个的子结构

思路：使用递归

判断两个树是否相等的三个条件是与的关系，即：

当前两个树的根节点值相等；
并且，s 的左子树和 t 的左子树相等；
并且，s 的右子树和 t 的右子树相等。
而判断 t 是否为 s 的子树的三个条件是或的关系，即：

当前两棵树相等；
或者，t 是 s 的左子树；
或者，t 是 s 的右子树。

作者：fuxuemingzhu
链接：https://leetcode-cn.com/problems/subtree-of-another-tree/solution/dui-cheng-mei-pan-duan-zi-shu-vs-pan-duan-xiang-de/
"""
from collections import deque

def isSubtree(s, t):
    """
    type s: TreeNode
    type t: TreeNode
    rtype: bool
    """
    if not s and not t:
        return True
    if not s or not t:
        return False
    return isSameTree(s, t) or isSameTree(s.left, t) or isSameTree(s.right, t)

def isSameTree(s, t):
    if not s and not t:
        return True
    if not s or not t:
        return False
    return s.val == t.val and isSameTree(s.left, t.left) and isSameTree(s.right, t.right)

class TreeNode(object):
    
    def __init__(self, x):
        self.val = x
        self.left = None
        self.right = None
        
class Tree(object):
    
    def __init__(self):
        self.root = None
        
    def construct_tree(self, values=None):
        if not values:
            return None
        self.root = TreeNode(values[0])
        queue = deque([self.root])
        leng = len(values)
        nums = 1
        while nums < leng:
            node = queue.popleft()
            if node:
                node.left = TreeNode(values[nums]) if values[nums] else None
                queue.append(node.left)
                if nums + 1 < leng:
                    node.right = TreeNode(values[nums + 1]) if values[nums + 1] else None
                    queue.append(node.right)
                    nums += 1
                nums += 1
    
    def bfs(self):
        ret = []
        queue = deque([self.root])
        while queue:
            node = queue.popleft()
            if node:
                ret.append(node.val)
                queue.append(node.left)
                queue.append(node.right)
        return ret
    
    def pre_traversal(self):
        ret = []
        def traversal(head):
            if not head:
                return
            ret.append(head.val)
            ret.append(head.left)
            ret.append(head.right)
        traversal(self.root)
        return ret
    
    def in_traversal(self):
        ret = []
        def traversal(head):
            if not head:
                return
            ret.append(head.left)
            ret.append(head.val)
            ret.append(head.right)
        traversal(self.root)
        return ret
    
    def post_traversal(self):
        ret = []
        def traversal(head):
            if not head:
                return
            ret.append(head.left)
            ret.append(head.right)
            ret.append(head.val)
        traversal(self.root)
        

if __name__ == '__main__':
    t1 = Tree()
    t1.construct_tree([1, 2, 3, 4, 5, 6, 7])
    print(t1.bfs())
    t2 = Tree()
    t2.construct_tree([2, 4, 5])
    print(t2.bfs())
    print(isSubtree(t1.root, t2.root))
        
```
