```python
"""
数组类型
题目地址：寻找数组的中心索引
题目来源：https://leetcode-cn.com/problems/find-pivot-index/
"""
class Solution:
    def pivotIndex(self, nums):  # (self, nums: List[int]) -> int:
        s = sum(nums)
        leftsum = 0
        for i, x in enumerate(nums):
            if leftsum == (s - leftsum - x):
                return i
            leftsum += x
        return -1
```

```python
"""
数组类型
题目名称：搜索插入的位置
题目来源：https://leetcode-cn.com/problems/search-insert-position/
"""
class Solution:
    def searchInsert(self, nums: List[int], target: int) -> int:
        # 我的方法
#         if target in nums:
#             return nums.index(target)
#         nums.append(target)
#         nums.sort()
#         return nums.index(target)
        # 大佬的方法
#         i = 0 
#         while i < len(nums) and nums[i] < target:
#             i += 1
#         return i
        # 题解二分法
        size = len(nums)
        if size == 0:
            return 0
        
        if nums[size - 1] < target:
            return size
    
        left, right = 0, size - 1
        while left < right:
            mid = (left + right) // 2
            if nums[mid] < target:
                left = mid + 1
            else:
                right = mid
        return left
```

```python
"""
数组类型
题目名称：合并区间
题目来源：https://leetcode-cn.com/problems/merge-intervals/
"""
class Solution:
    def merge(self, intervals: List[List[int]]) -> List[List[int]]:
        intervals.sort(key = lambda x: x[0])
        ret = []
        for i in intervals:
            if not ret or ret[-1][1] < i[0]:
                ret.append(i)
            else:
                ret[-1][1] = max(ret[-1][1], i[1])
        return ret  
```

```python
"""
二维数组类型
题目名称：旋转矩阵
题目来源：https://leetcode-cn.com/problems/rotate-matrix-lcci/
"""
class Solution:
    def rotate(self, matrix: List[List[int]]) -> None:
        # 方法一：内置函数无敌
        matrix[::] = zip(*matrix[::-1])  
        
        # 方法二：水平翻转，再主对角线翻转
        n = len(matrix)
        for i in range(n // 2):
            for j in range(n):
                matrix[i][j], matrix[n-i-1][j] = matrix[n-i-1][j], matrix[i][j]
        for i in range(n):
            for j in range(i):
                matrix[i][j], matrix[j][i] = matrix[j][i], matrix[i][j]
```

```python
"""
二维数组类型
题目名称：零矩阵
题目来源：https://leetcode-cn.com/problems/zero-matrix-lcci/
"""
class Solution:
    def setZeroes(self, matrix: List[List[int]]) -> None:
        row, col, row_list, col_list = len(matrix), len(matrix[0]), set(), set()
        for i in range(row):
            for j in range(col):
                if matrix[i][j] == 0:
                    row_list.add(i)
                    col_list.add(j)
                    
        for i in row_list:
            for j in range(col):
                matrix[i][j] = 0 
                
        for i in range(row):
            for j in col_list:
                matrix[i][j] = 0
```

```python
"""
字符串简介
题目名称：最长公共前缀
题目来源：https://leetcode-cn.com/problems/longest-common-prefix/
"""
class Solution:
    def longestCommonPrefix(self, strs: List[str]) -> str:
        # zip，set内置函数特性
        res = ""
        for tmp in zip(*strs):
            tmp_set = set(tmp)
            if len(tmp_set) == 1:
                res += tmp[0]
            else:
                break
        return res
    
        # 方法二：最大最小前缀
        if not strs:
            return ""
        s1, s2 = min(strs), max(strs)
        for i, x in enumerate(s1):
            if x != s2[i]:
                return s2[:i]
        
        return s1
```

```python
"""
字符串简介
题目名称：最长回文子串
题目来源：https://leetcode-cn.com/leetbook/read/array-and-string/conm7/
"""
class Solution:
    def longestPalindrome(self, s: str) -> str:
    # 方法一：动态规划 看不懂
        size = len(s)
        if size < 2:
            return s
        
        dp = [[False for _ in range(size)] for _ in range(size)]
        
        max_len, start = 1, 0
        
        for i in range(size):
            dp[i][i] = True
            
        for j in range(1, size):
            for i in range(0, j):
                if s[i] == s[j]:
                    if j - i < 3:
                        dp[i][j] = True
                    else:
                        dp[i][j] = dp[i + 1][j - 1]
                else:
                    dp[i][j] = False
                
                if dp[i][j]:
                    cur_len = j - i + 1
                    if cur_len > max_len:
                        max_len = cur_len
                        start = i
        return s[start: start + max_len]
    
        
    # 方法二：以任意节点为中心，向两边发散，看其是否是回文子串。
        size = len(s)
        if size < 2:
            return s
        max_len = 1
        res = s[0]
        
        for i in range(size):
            palindrome_odd, odd_len = self.__center_spread(s, size, i, i)  # 奇数
            palindrome_even, even_len = self.__center_spread(s, size, i, i + 1)  # 偶数

            # 当前找到的最长回文子串
            cur_max_sub = palindrome_odd if odd_len >= even_len else palindrome_even
            if len(cur_max_sub) > max_len:
                max_len = len(cur_max_sub)
                res = cur_max_sub

        return res

    def __center_spread(self, s, size, left, right):
        """
        left = right 的时候，此时回文中心是一个字符，回文串的长度是奇数
        right = left + 1 的时候，此时回文中心是一个空隙，回文串的长度是偶数
        """
        i = left
        j = right

        while i >= 0 and j < size and s[i] == s[j]:
            i -= 1
            j += 1
        return s[i + 1:j], j - i - 1
```

```python
"""
字符串简介
题目名称：翻转字符串里的单词
题目来源：https://leetcode-cn.com/problems/reverse-words-in-a-string/
"""
class Solution:
    def reverseWords(self, s: str) -> str:
        s = s.strip()
        i = j = len(s) - 1
        res = []
        while i >= 0:
            while i >= 0 and s[i] != " ":
                i -= 1
            res.append(s[i + 1: j + 1])
            while s[i] == " ":
                i -= 1
            j = i
        return " ".join(res)
```

```python
"""
双指针技巧-双向指针
题目名称：反转字符串
题目来源：https://leetcode-cn.com/problems/implement-strstr/
"""
class Solution:
    def reverseString(self, s: List[str]) -> None:
        """
        Do not return anything, modify s in-place instead.
        """
        i, j = 0, len(s) - 1
        while i < j:
            s[i], s[j] = s[j], s[i]
            i += 1
            j -= 1
        
```

```python
"""
双指针技巧-双向指针
题目名称：两数之和II - 输入有序数组
题目来源：https://leetcode-cn.com/problems/two-sum-ii-input-array-is-sorted/
"""
class Solution:
    def twoSum(self, numbers: List[int], target: int) -> List[int]:
        left, right = 0, len(numbers) - 1
        while left < right:
            if numbers[left] + numbers[right] == target:
                return [left + 1, right + 1]
            elif numbers[left] + numbers[right] < target:
                left = left + 1
            else:
                right = right - 1
```

```python
"""
双指针技巧-快慢指针
题目名称：移除元素
题目来源：https://leetcode-cn.com/problems/remove-element/
"""
class Solution:
    def removeElement(self, nums: List[int], val: int) -> int:
        try:
            while True:
                nums.remove(val)
        except:
            return len(nums)
        
        # 方法二
        slow = 0
        n = len(nums)
        for fast in range(n):
            if nums[fast] != val:
                nums[slow] = nums[fast]
                slow += 1
        return slow
```

```python
"""
双指针技巧-快慢指针
题目名称：长度最小的子数组
题目来源：https://leetcode-cn.com/problems/minimum-size-subarray-sum/
"""
class Solution:
    def minSubArrayLen(self, s: int, nums: List[int]) -> int:
        if s > sum(nums):
            return 0
        left, right, res, sum_lr = 0, 0, len(nums) + 1, 0
        while right < len(nums):
            while sum_lr < s and right < len(nums):
                sum_lr += nums[right]
                right += 1
            while sum_lr >= s and left >= 0:
                res = min(res, right - left)  # 此处 right - left 没有 + 1  是因为 上面的 while 操作中 已经加过了
                sum_lr -= nums[left]
                left += 1
        return res
```

```python
"""
小结
题目名称：杨辉三角
题目来源：https://leetcode-cn.com/problems/pascals-triangle/

这道题有必要写一个题解，给 20200812 之后的叶看，来自于题解
[
     [1],
    [1,1],
   [1,2,1],
  [1,3,3,1],
 [1,4,6,4,1]
]


当前一行只比上一行多一个元素，关键点在于：本行元素等于上一行元素，元素错位相加。如下:

1  3  3  1  0
0  1  3  3  1
===============
1  4  6  4  1


因此，我们只要对最后一行单独处理，最后一行行首，行尾分别添加一个零，然后对应位置求和即可得到新的一行。
"""
class Solution:
    def generate(self, numRows: int) -> List[List[int]]:
        if numRows == 0:
            return []
        res = [[1]]
        while len(res) < numRows:
            newRow = [a + b for a, b in zip([0] + res[-1], res[-1] + [0])]
            res.append(newRow)
        return res
```

```python
"""
小结
题目名称：杨辉三角II
题目来源：https://leetcode-cn.com/problems/pascals-triangle-ii/
"""
class Solution:
    def getRow(self, rowIndex): # (self, rowIndex: int) -> List[int]:
        if rowIndex == 0:
            return [1]
        res = [1]
        while len(res) < rowIndex + 1:
            res = [a + b for a, b in zip([0] + res, res + [0])]
        return res
    
if __name__ == '__main__':
    solution = Solution()
    print(solution.getRow(1))
```

```python
"""
小结
题目名称：寻找旋转排序数组中的最小值
题目来源：https://leetcode-cn.com/problems/reverse-words-in-a-string-iii/
大佬题解：https://leetcode-cn.com/problems/find-minimum-in-rotated-sorted-array/solution/er-fen-cha-zhao-wei-shi-yao-zuo-you-bu-dui-cheng-z/
"""
class Solution:
    def findMin(self, nums: List[int]) -> int:
        left, right = 0, len(nums) - 1
        while left < right:
            mid = (left + right) // 2   # 也可以写成 (left + right) >> 1
            if nums[mid] > nums[right]:
                left = mid + 1
            else:
                right = mid
        return nums[left]
```

```python
"""
小结
题目名称：寻找旋转排序数组中的最小值 II  有重复数值
题目来源：https://leetcode-cn.com/problems/find-minimum-in-rotated-sorted-array-ii/
大佬题解：https://leetcode-cn.com/problems/find-minimum-in-rotated-sorted-array-ii/solution/154-find-minimum-in-rotated-sorted-array-ii-by-jyd/
"""
class Solution:
    def findMin(self, nums: List[int]) -> int:
        left, right = 0, len(nums) - 1
        while left < right:
            mid = (left + right) >> 1
            if nums[mid] > nums[right]:
                left = mid + 1
            elif nums[mid] < nums[right]:
                right = mid
            else:
                right -= 1
        return nums[left]
```

```python
"""
小结
题目名称：删除排序数组中的重复项
题目来源：https://leetcode-cn.com/problems/remove-duplicates-from-sorted-array/
"""
class Solution:
    def removeDuplicates(self, nums: List[int]) -> int:
        if not nums: return 0
        i = 0 
        for j in range(1, len(nums)):
            if nums[i] != nums[j]:
                i += 1
                nums[i] = nums[j]
        return i + 1
```

```python
"""
小结
题目名称：移动零
题目来源：https://leetcode-cn.com/problems/move-zeroes/
"""
class Solution:
    def moveZeroes(self, nums: List[int]) -> None:
        """
        Do not return anything, modify nums in-place instead.
        """
        if not nums:
            return []
        i = 0
        for j in range(len(nums)):
            if nums[j] != 0:
                nums[i] = nums[j]
                i += 1
        for i in range(i, len(nums)):
            nums[i] = 0
        return nums
```

```python
"""
小结
题目名称：两数之和
题目来源：https://leetcode-cn.com/problems/two-sum/
"""
class Solution:
    def twoSum(self, nums: List[int], target: int) -> List[int]:
        hashmap = {}
        for index, num in enumerate(nums):
            another_num = target - num
            if another_num in hashmap:
                return [hashmap[another_num], index]
            hashmap[num] = index
        return None
```
